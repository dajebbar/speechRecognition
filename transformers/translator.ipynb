```python
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    pipeline
)

model = AutoModelForSeq2SeqLM.from_pretrained(
      "facebook/nllb-200-distilled-600M"
)

tokenizer = AutoTokenizer.from_pretrained(
    "facebook/nllb-200-distilled-600M"
)
  
translator = pipeline(
    task="translation",
    model=model,
    tokenizer=tokenizer,
    src_lang="eng_Latn",
    tgt_lang="fra_Latn",
    max_length=300
    
    )
```

```python
def translate_text(file):
    text_tr = ""
    with open(file, "r") as f:
      content_size = 100
      f_content = f.read(content_size)
      while len(f_content) > 0:
        text = f_content
        f_content = f.read(content_size)
        # print(text)
        text_tr += translator(text)[0]["translation_text"] + "\n"
    with open("tr.txt", "a") as f_tr:
      f_tr.write(text_tr)

    return

file = "pdf.txt"
translate_text(file)
```
