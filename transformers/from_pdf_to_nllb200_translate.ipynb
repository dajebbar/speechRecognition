```python
import os
from pdfminer.high_level import extract_text
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    pipeline
)

def get_text(pth):
  # os.makedirs("origin_files", exist_ok=True)
  os.makedirs("files", exist_ok=True)
  for i in pth:
    text  = extract_text(i)
    with open(f'./files/{i}.txt', 'a') as f:
      f.write(text)
  return "text to translate is ready"
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained(
      "facebook/nllb-200-distilled-600M"
)

tokenizer = AutoTokenizer.from_pretrained(
    "facebook/nllb-200-distilled-600M"
)
  
translator = pipeline(
    task="translation",
    model=model,
    tokenizer=tokenizer,
    src_lang="eng_Latn",
    tgt_lang="fra_Latn",
    max_length=10000
    
    )

def translate_text(pth):
  os.makedirs("xfile_tr", exist_ok=True)

  for i, item in enumerate(pth, 1):
    with open(item, "r") as f:
      content_size = 50
      f_content = f.read(content_size)
      while len(f_content) > 0:
        text = f_content
        f_content = f.read(content_size)
        text_tr = translator(text)[0]["translation_text"]
        print(f"Text {i+1}:\n")
        print(text_tr)
        print("*-*" * 15)
    # with open(os.path.join("./xfile_tr", f"file_{i}.txt"), "a") as f_tr:
    #     f_tr.write(text_tr)
    return "Done!"


```
